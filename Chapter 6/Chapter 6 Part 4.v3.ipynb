## 6.10 Continuous fermentation

Most bulk scale chemical processes are operated in continuous mode, consider for example the traditional petrochemical refinery. Continuous operation offers the advantage of stability, where conditions remain the same over extended periods of time. The non-production time in expensive equipment is minimised with continuous production, while optimal conditions can be targeted at steady state.

Fermentation processes can also operate at steady state. Although commercial continuous processes exist, the status quo is batch and fed-batch operation. It is much easier to operate a batch fermentation, especially in the lab, where small scale explorations establish the platform for commercial development. Unfortunately people tend to stick with what works and thus the success from the lab scale batch-experiment is often merely copied when a commercial process is developed. This is likely to change in the age of the biorefinery, where optimisation of bulk scale fermentation processes will become more important.

Continuous fermentation holds some practical challenges.Long term sterility can be an issue, where infections from unwanted organisms can kill a process and cause major material losses. Genetic stability of modified microbes can also be an issue, where strain regression can occur over extended periods of operation. With proper engineering these challenges can be overcome.

In the lab continuous fermentation has advantages over conventional batch fermentation. The main advantage is that the microbe can be operated at steady state by keeping all the conditions in the fermenter constant. It is much easier to understand and investigate the physiology of an organism if it is ‘standing still’. In a batch system conditions are always changing and it becomes harder to investigate the effect of the conditions on the organism. Unfortunately continuous fermentation on small scale require specialized equipment, where a batch fermentation can be as easy as putting a flask in a temperature controlled shaker, accordingly 95% of fermentation literature is based on batch of fed-batch fermentation.   

This section takes you through the modelling aspects of continuous fermentation with the aim of showing you the advantages of using this mode of operation. It will be shown that differential equations as well as algebraic equations can be used to solve the steady state models. In addition a small amount of the section will be dedicated to cell retainment, a very useful strategy when continuous conditions are employed. We’ll once again explain the principles via example. 

Let's define the properties of the microbe that we will be using for example. We'll neglect substrate inhibition, but include product inhibition. The following stoichiometry is assumed for the anaerobic system (cmol yields):

 |$Y_{SX}$|$Y_{SP}$|$Y_{S ATP}$
 -|----|-----|--
 **Growth**|0.12|0.6|0.38
 **Maintenace**|0|0.75|0.46
 
The inhibition parameters are given by:

 |$K_M \, \left(\frac{cmol}{L} \right)$|$K_P \, \left(\frac{cmol}{L} \right)$
 -|-|-
 **Growth**|0.0005|0.5
 **Mainetnace**|0.0005|$\infty$
 
 Product inhibition is given by the follwoing function:
 
 $$f_{PI}= \left[1+\frac{C_P}{K_P} \right]^{-1} $$
 
 The maximum physiological rates are given by:
 
 $\mu_{max}\,\left(\frac{1}{h} \right)$|$\theta_{max} \, \left(\frac{mol\, ATP}{cmol\, X\cdot h} \right) $
 -|-
 0.3 |0.14 
 
 We can now define the parameters in Python. We'll start with a $10 m^3$ batch reactor where $C_S^o=3.5 \frac{cmol}{L}$ and $C_X^o=0.0003 \frac{cmol}{L}$.

```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline 
from scipy.integrate import odeint

mumax,thetamax, Km, Kp= 0.3, 0.14, 0.0005, 0.5
Cxo, Cso = 0.0003, 3.5
Yg=[1, 0.12, 0.6 , 0.38] 
Ym=[1, 0, 0.75, 0.46]                     
Vo=10000                                   
Co=numpy.array([Cxo, Cso, 0, 1])     #[X, S, P, V]             
No=Co*Vo
Qf=Q=Cxf=Csf=Cpf=0                    #For now it is a batch fermenter
```

```python
def r_prime(C):
    Cx, Cs, Cp = C          
    mu=mumax*Cs/(Km+Cs)*(1+Cp/Kp)**(-1)
    theta=thetamax*Cs/(Km+Cs)
    
    return [mu,
            -((Yg[0]/Yg[1])*mu+(Ym[0]/Ym[3])*theta),     
            (Yg[2]/Yg[1])*mu+(Ym[2]/Ym[3])*theta]
```

```python
def dNdt_fun(N,t):
    Cx, Cs, Cp, V = N[0]/N[3],N[1]/N[3],N[2]/N[3],N[3]  #calculating and naming concentration and volume 
    r=r_prime([Cx, Cs, Cp])
    
    return [Qf*Cxf-Q*Cx+(r[0])*Cx*V,
            Qf*Csf-Q*Cs+(r[1])*Cx*V,    
            Qf*Cpf-Q*Cp+(r[2])*Cx*V,                        
            Qf-Q] 
```

Similar to chapter 6 part 3 we'll define a seperate function to obtain and plot the fermentation results:


```python
def solve_ferm(tmax):
    tspan=numpy.linspace(0,tmax,200)             #define the timespan of the integration
    N = odeint(dNdt_fun, No, tspan)

    Cx=N[:,0]/N[:,3]                           #devide cmol amount by the volume to get concentration 
    Cs=N[:,1]/N[:,3]
    Cp=N[:,2]/N[:,3]
    V=N[:,3]

    plt.plot(tspan, Cp, color='blue', label='p')
    plt.plot(tspan, Cs, color='red',label='s')
    plt.plot(tspan, Cx, color='green',label='x')
    plt.legend(loc='best')
    plt.ylabel('Concentration cmol/L') 
    plt.xlabel('time (h)') 
    plt.show()

    return N,tspan
```

Let's se what the batch run will deliver over 50 hours:

```python
N1, tspan = solve_ferm(50)
```

Let's calculate the productivity and yield when the fermentation is completed:

```python
Cs1=N1[:,1]/N1[:,3]   #first integration
tdone1=numpy.interp(0.001,Cs1[::-1],tspan[::-1])  #tdone is time when substrate is depleted
Cp1=N1[:,2]/N1[:,3]   
P1=max(Cp1)/tdone1   #productivity of first integration
Ysp_o=max(Cp1)/Cso
[tdone1, P1, Ysp_o]
```

We will now repeat the fermentation above, but his time we’ll allow for a constant flow through the fermenter ($Q^{\,f}=Q$). We’ll only feed substrate in the feed ($C_X^f=C_P^f=0$), but note that biomass, product and substrate will be removed in the outlet flow. Also note that the volume won’t change since $Q^{\,f}=Q$. We’ll start by using $Q^{\,f}=1000 \frac{L}{h}$ ($\frac{1}{10}$ of the volume added (and removed) each hour) and $C_S^f=3.5 \frac{cmol}{L}$, while no substrate will initially be present in the fermenter ($C_S^o=0$). We will however start with the inoculum amount in the fermenter ($C_X^o=0.0003 \frac{cmol}{L}$).

```python
Cso=0;
Co=numpy.array([Cxo, Cso, 0, 1])     #[X, S, P, V]             
No=Co*Vo
Qf=Q=1000                            # define throughflow 
Csf=3.5                              # substrate in feed 
Cxf=Cpf=0                            # zero product and cells in feed line
```

Let's see what happens in our first continuous fermentation:

```python
N2, tspan2 = solve_ferm(150)
```

There is an overshoot in $C_S$ but steady state (no change) is reached after about 100 hours. The substrate overshoot can be related to the small amount of biomass initially present in the fermenter. Interesting that a steady state biomass concentration ($C_X$) is reached since biomass is continuously removed from the fermenter. Let's calculate how much biomass we are losing per hour:

```python
Cx2=N2[:,0]/N2[:,3]
Cx_out=Cx2[-1]                       #Outlet concnetration at the end
Cells_out=Cx_out*Q                   #cmol/h
Cells_out_mass=Cells_out*24.6/1000   #kg/h 
Cells_out_mass
```

Given a biomass cmol weight of $24.6 \frac{g}{cmol}$ we are losing $3.38 kg$ of biomass every hour. Where is this biomass coming from? From the fermenter and nowhere else is the answer, so it is clear that we are making (growing) $3.38 kg$ of cells every hour in this $10 m^3$ fermenter. What will happen is we lower the througput (smaller $Q^{\,f}$)?

```python
Qf=Q=500
N3, tspan3 = solve_ferm(140)
```

We get to steady state after 60+ hours of 'start-up' and all the substrate (almost all) is consumed in the fermenter. Accordingly $C_P$ and $C_S$ are higher than before. It will be interesting to compare the productivity of the steady state to the batch results obtained in simulation 1 (N1). Remember that productivity is defined as the amount of product per time per volume of fermenter, so:

$$P_{cont}=\frac{C_P^{out}Q^{\,f}}{V}$$

```python
Cp3=N3[:,2]/N3[:,3]
Cs3=N3[:,1]/N3[:,3]
Cx3=N3[:,0]/N3[:,3]

P3=max(Cp3)*Q/Vo
[P3/P1]
```

So the productivity of the continuous fermenter is almost double than that of the batch fermenter. We have more efficient use of expensive space. The main reason will be the large amount of biomass ($C_X$) in the fermenter at steady state. We can neglect the 'start-up' time where biomass was accumulating since the fermenter will be running for numerous hours at steady state (>1000 hours). Play around with the throughput ($Q$) and investigate the change in fermenter profiles and productivity. 

From the results it is clear that steady state is attained where the change in concentrations are zero. Since the volume is constant the change in cmol amounts will also be zero. The ever present cmol balance thus reduces to:

$$\frac{dN_i}{dt}=Q^F C^F_i-QC_i  + r^{'}_i C_X V =0 $$  

With the differential term out of the equation we are now left with a set of algebraic equations. Let's sovle for the steady state above by not using the differential equation. We'll slightly alter the DNdt_fun in order to only have the 3 cmol balances. We include Q as an input since we will vary the throughput at a later stage. 

```python
def steady_state(C,Q):
    Cx, Cs, Cp = C               #name assignment 
    r=r_prime([Cx, Cs, Cp])
    Qf=Q
    
    return [Qf*Cxf-Q*Cx+(r[0])*Cx*Vo,
            Qf*Csf-Q*Cs+(r[1])*Cx*Vo,    
            Qf*Cpf-Q*Cp+(r[2])*Cx*Vo]                        
             
```

Given this 3 algebraic equations, we need to solve for a combination of $C_X, C_S, C_P$ that satisfies the 3 equations. Note that the same microbial response function (**r_prime**) is used in the cmol balances. We'll use the function **fsolve** to perform the root finding. Note that the function **fsolve** requires an estimate for the root, preferably not that far from the root itself. We are not interested in solving $Q$ since it is specified and we therefore pass the set value of $Q$ to the steady_state function via the args handle. 

```python
from scipy.optimize import fsolve
C_steady=fsolve(steady_state,[0.5,0,2.5],args=Q)  #Get estimate from graph above
C_steady
```

Let's compare the answer to the steady state achieved via the differential equation:

```python
C_fromDE=[Cx3[-1], Cs3[-1], Cp3[-1]]
C_fromDE,[2,2,2]
```

Almost identical answers. The small difference can be attributed to the fact that the D.E. answer was not exactly at steady state after 140 hours. Run the DE for a longer time and observe how the difference gets even smaller.

It will be interesting to determine how the steady state varies for different throughputs. We can run **fsolve** numerous times for different values of $Q$.

```python
Qspan=numpy.linspace(200,2500,20)
C=[]                                               #start with an empty matrix
for Qi in Qspan:
    Co = fsolve(steady_state, [1,1,1], args=Qi)    #solve for specific Qf 
    C.append(Co)                                   #add (solution) rows to the C matrix

plt.plot(Qspan, C,'o')
plt.legend(['Cx', 'Cs', 'Cp'], loc='best')
plt.ylabel('Concentration cmol/L') 
plt.xlabel('Q (L/h)') 
plt.show() 
```

You will see that the first two fsolve solutions (low $Q$) returned strange roots. It is important to always check the fsolve answers to see if they are physically realisable:

$$0<C_S \le C_S^o$$

$$0<C_P \le Y_{SP}^M \cdot C_S^o$$  

$$0<C_P \le Y_{SX}^G \cdot C_S^f $$   

These criteria only apply when biomass is not retianed. For multiple fsolve loops it is a good idea to use the previous answer as the guess for the next: 

```python
Qspan=numpy.linspace(200,2500,20)
C=[]
Co = [0.4,0,2.2]                            #initial guess for the first fsolve procedure
for Qi in Qspan:
    Co = fsolve(steady_state, Co, args=Qi)  #Previous answer Co used as guess
    C.append(Co)

plt.plot(Qspan, C,'o')
plt.legend(['Cx', 'Cs', 'Cp'], loc='best')
plt.ylabel('Concentration cmol/L') 
plt.xlabel('Q (L/h)')
plt.show() 
```

Much better. Full substrate conversion is achieved for low flowrates, while very little reaction occurs at high throughput (large $Q$). Note that the graph above represents the results from **10 different steady states (separate runs).** 

The continuous fermenter is referred to as the **chemostat**. The traditional variable for quantifying throughput is defined as follow:

$$ D= \frac{Q}{V}$$
-------------

With D referred to as **dilution rate** with units $\frac{1}{h}$. Let's replot the chemostat results against D:

```python
D=Qspan/Vo

plt.plot(D, C,'o')
plt.legend(['Cx', 'Cs', 'Cp'], loc='best')
plt.ylabel('Concentration cmol/L') 
plt.xlabel('D (1/h)')
plt.show()
```

By manipulation of the biomass cmol balance the following is obtained:

$$Q C^F_X-QC_X  + r^{'}_i C_X V =0 $$  

$$QC_X = \mu C_X V  $$ 

$$D = \mu   $$ 

This is very interesting! The result implies that the **growth rate is determined by the throughput**. So at low $D$ the growth rate will be low. It makes sense since $C_S$ is low (Monod) and $C_P$ is high (product inhibition) at low $D$. Despite this $C_X$ is at it highest, but this is because the removal of cells from the fermenter is slow due to the small $Q$. At high $D$, $C_S$ is high ($f_M \approx 1)$ and $C_P$ is low and thus $\mu$ has to be high.

But what will happen if $D > \mu_{max}$? Let's see:


```python
D1=0.35
Q=D1*Vo
C_steady=fsolve(steady_state,[0.1,3,0.1],args=Q)  
C_steady
```

The answer is absolutely nothing! The feed will be identical to the outlet. The condition is referred to as 'washout' since all biomass is washed out of the fermenter. So always make sure that:

$$D<\mu_{max}$$

Let's have a look at productivity as a function of $D$:

```python
C1=numpy.array(C)
Cp=C1[:,2]
P=Cp*D

plt.plot(D, P,'o')
plt.ylabel('Productivity cmol/(L.h)') 
plt.xlabel('D (1/h)')
plt.show()
```

The productivity clearly goes through a maximum. The maximum is due to a trade-off between $\mu$ and $C_X$ (the one is low where the other is high). We can also plot the observed yields as a function of $D$:

```python
Cs=C1[:,1]
Ysp_o=Cp/(Csf-Cs)

plt.plot(D, Ysp_o,'o')
plt.ylabel('Product yield cmol/cmol') 
plt.xlabel('D (1/h)')
plt.show()
```

Explain the higher yield at low $D$?

### Cell retainment

The volumetric rate of product formation ($r_p$) is a function of the biomass concentration in the fermenter ($C_X$). In order to increase $C_X$ it is not a bad idea to separate cells from the outlet stream and recycle them back to the fermenter. We won't be going into the details but equipment like hydrocyclones and tangential filters are able to perform this separation in practise. A bleed stream ($Q^{\,b}$) is typically included when cell separation is performed in order to counter excessive cell accumulation . The bleed flowrate is set to be 5-20% of the inlet flowrate, while the outlet flowrate from the separation ($Q^{\,sep}$) is adjusted to keep the total flowrate out of the fermenter the same as the incoming flowrate. $C_S$ and $C_P$ in $Q^{\,b}$ and $Q^{\,sep}$ will be the same as in the fermenter, while the $C_X$ in $Q^{\,sep}$ will be zero and $C_X$ in $Q^{\, b}$ will be the same as in the fermenter.

<img src="retain1.png" width="500" />

Make sure you agree with the following modification of the cmol balance. The parameter $f_b$ is merely defined by:

$$Q^{\,b}=f_{b} \cdot Q^{\, f}=f_{b} \cdot (Q^{\,b}+Q^{\,sep})$$

```python
def steady_state(C,Q):
    Cx, Cs, Cp = C               #name assignment 
    r=r_prime([Cx, Cs, Cp])
    Qf=Q
    
    return [Qf*Cxf-Q*fbleed*Cx+(r[0])*Cx*Vo,
            Qf*Csf-Q*Cs+(r[1])*Cx*Vo,    
            Qf*Cpf-Q*Cp+(r[2])*Cx*Vo]             
```

We can investigate the steady state performance like before. Note that we are not restricted by $D<\mu_{max}$ anymore, since the cells are retained in the fermenter. 

```python
fbleed=0.2                              #fraction of outlet flow as bleed stream
Qspan=numpy.linspace(500,8000,20)
D=Qspan/Vo
C=[]
Co = [0.4,0,2.2]                            #initial guess for the first fsolve procedure
for Qi in Qspan:
    Co = fsolve(steady_state, Co, args=Qi)  #Previous answer Co used as guess
    C.append(Co)

plt.plot(D, C,'o')
plt.legend(['Cx', 'Cs', 'Cp'], loc='best')
plt.ylabel('Concentration cmol/L') 
plt.xlabel('D (1/h)')
plt.show() 
```

Play around with different $f_b$ values. Let's plot the volumetric productivity:

```python
C2=numpy.array(C)
Cp2=C2[:,2]
P=Cp2*D

plt.plot(D, P,'o')
plt.ylabel('Productivity cmol/(L.h)') 
plt.xlabel('D (1/h)')
plt.show()
```

Can you see the major improvement? The question now becomes how the cost of seperation equipment influences the economics. The fermentation vessel is 5 times more productive, but also much more expensive.
