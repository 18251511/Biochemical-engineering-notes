## 6.9 Fed-batch fermentation

A fed-batch fermeter is where an inlet is added to the fermenter, while nothing is removed from the fermenter. Accordingly there is a build-up of volume. The main question to address is why this strategy is so often followed in industry. This section will attempt to answer the question via an example.

Before starting with the example we once again want to return to the main rate parameters $\mu$ and $\theta$ and the influence of the external environment on these parameters. The generic functions of $\mu$ and $\theta$ are given by the following:

$$\mu=\mu_{max} \cdot f_M^{\mu} \cdot f_{PI}^{\mu} \cdot f_{SI}^{\mu}$$

$$\theta=\theta_{max} \cdot f_M^{\theta} \cdot f_{PI}^{\theta} \cdot f_{SI}^{\theta}$$

The generic form of the equation is given by the maximum value multiplied by 3 fractions or inhibition functions ($f$). It is very important to understand that all the inhibition fractions will **always vary between zero and unity**. If the function value remains unity it will have no influence on the $\mu$ and $\theta$ functions. The $f_M$ functions are the Monod term and should always be used. These functions only become less than unity at low subtrate concentrations and becomes zero when all the substrate is depleted. Discarding these functions will result in unrealistic model results (see chapter 6 part 2). In addition the substrate depletion function ($f_M$) will the same for $\mu$ and $\theta$ such that: $f_M^{\mu}=f_M^{\theta}$.

The product inhibition ($f_{PI}$) fractions or functions have been used in chapter 6 part 2. Typical formulations for $f_{PI}$ include:

$$f_PI= \left[1-\frac{C_P}{C_P^*} \right] $$

or 

$$f_PI= \left[1+\frac{C_P}{K_P} \right]^{-1}  $$

Note that the termination model will result in complete inhibition ($f_{PI}=0$) if the critical product concentration is reached ($C_P=C_P^*$). Also note that $f_{PI}^{\mu} \neq f_{PI}^{\theta}$. It typically happens that growth stops ($f_{PI}^{\mu}=0$) while maintenance is still active ($ 0<f_{PI}^{\theta} \le 1$). For this scenario $f_{PI}^{\mu}$ will result in a negative fraction and correction via an if statement is required (see section 6.8) since the inhibition rule is always: $ 0\le f \le 1$.

It is also possible for substrate to inhibit the $\mu$ and $\theta$ functions. The following function is commonly used for substrate inhibition:

$$f_{SI}= \left[1+\frac{C_S}{K_S} \right]^{-1}  $$

Similar to product inhibition $f_{SI}^{\mu} \neq f_{SI}^{\theta}$. This will imply different paramater values of $K_S$ for $f_{SI}^{\mu}$ and $f_{SI}^{\theta}$. Note that the absense of substrate inhibition is given by $f_{SI}^{\mu}=f_{SI}^{\theta} =1$.

Let's get on with the example. We'll assume negligible product inhibition for this example ($f_{PI}^{\mu} = f_{PI}^{\theta} =0$). The maintenance and growth stoichiometry are given in the table below (cmol yields):

 |$Y_{SX}$|$Y_{SP}$|$Y_{S ATP}$
 -|----|-----|--
 **Growth**|0.15|0.65|0.25
 **Maintenace**|0|0.85|0.33
 
The inhibition parameters are given by:

 |$K_M \, \left(\frac{cmol}{L} \right)$|$K_P \, \left(\frac{cmol}{L} \right)$|$K_S \,  \left(\frac{cmol}{L} \right)$
 -|-|-|-
 **Growth**|0.0005|$\infty$|2
 **Mainetnace**|0.0005|$\infty$|$\infty$
 
 The maximum physiological rates are given by:
 
 $\mu_{max}\,\left(\frac{1}{h} \right)$|$\theta_{max} \, \left(\frac{mol\, ATP}{cmol\, X\cdot h} \right) $
 -|-
 0.4 |0.05 
 
 We can now define the parameters in Python. We'll start with a batch reactor where $C_S^o=3 \frac{cmol}{L}$ and $C_X^o=0.0007 \frac{cmol}{L}$.

```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline 
from scipy.integrate import odeint

mumax,thetamax, Km, Ks= 0.4, 0.05, 0.0005, 2
Cxo, Cso = 0.0007, 3
Yg=[1, 0.15, 0.65 , 0.25] 
Ym=[1, 0, 0.85, 0.33]                     
Vo=4000                                   
Co=numpy.array([Cxo, Cso, 0, 1])     #[X, S, P, V]             
No=Co*Vo
```

We can now define the microbial response function (r_prime)

```python
def r_prime(C):
    Cx, Cs, Cp = C          
    mu=mumax*Cs/(Km+Cs)*(1+Cs/Ks)**(-1)
    theta=thetamax*Cs/(Km+Cs)
    
    return [mu,
            -((Yg[0]/Yg[1])*mu+(Ym[0]/Ym[3])*theta),     
            (Yg[2]/Yg[1])*mu+(Ym[2]/Ym[3])*theta]
```

Set all inlet and outlet flowrates to zero (batch fermenter)


```python
Qf=Q=Cxf=Csf=Cgf=Cef=0                    #For now it is a batch fermenter
```

Don't forget the differential equation (D.E.), the magic link between the microbe response and fermenter concentration profiles:

```python
def dNdt_fun(N,t):
    Cx, Cs, Cp, V = N[0]/N[3],N[1]/N[3],N[2]/N[3],N[3]  #calculating and naming concentration and volume 
    r=r_prime([Cx, Cs, Cp])
    
    return [Qf*Cxf-Q*Cx+(r[0])*Cx*V,
            Qf*Csf-Q*Cs+(r[1])*Cx*V,    
            Qf*Cgf-Q*Cp+(r[2])*Cx*V,                        
            Qf-Q] 
```


Let's integrate and look at the response. But unlike before we are going to put the integration and plot procedure into a function to reduce the copy and pasting of code required when re-running the integration. The function will plot the concentration profiles and return the output (N) and timespan (tspan) from the integration. The total fermentation time need to be specified in this function (tmax).

```python
def solve_ferm(tmax):
    tspan=numpy.linspace(0,tmax,200)             #define the timespan of the integration
    N = odeint(dNdt_fun, No, tspan)

    Cx=N[:,0]/N[:,3]                           #devide cmol amount by the volume to get concentration 
    Cs=N[:,1]/N[:,3]
    Cp=N[:,2]/N[:,3]
    V=N[:,3]

    plt.plot(tspan, Cp, color='blue', label='p')
    plt.plot(tspan, Cs, color='red',label='s')
    plt.plot(tspan, Cx, color='green',label='x')
    plt.legend(loc='best')
    plt.ylabel('Concentration cmol/L') 
    plt.xlabel('time (h)') 
    plt.show()

    return N,tspan
```

```python
N1, tspan = solve_ferm(40)
```

It appears that the fermentatation struggled to get going. This might be linked to the substrate inhibition. Let's repeat the fermentation by neglecting substrate inhibition. We can do this by specifying a large numerical value for $K_S$ that approximates infinity (thus $f_{SI}^\mu \approx 1$).

```python
Ks=1e10
N2,tspan =solve_ferm(40)  #re-running the fermentation model with zero substrate inhibition
```

Much better and much faster! The substrate inhibition is hampering with the productivity (economics) of the process. What exactly is productivity? It is merely the amount of product generated per active volume per time. For batch or fed batch processes we can take the final volume as the active volume and thus only devide the final product concentration by the time it took to achieve the product concentration:

$$P_{batch}=P_{fed-batch}=\frac{C_P^{final}}{t^{final}}$$

In order to accurately obtain the time where the substrate is depleted, we need to interpolate using the numpy function interp. The statement below finds the time where $C_S=0.001$ by working backwards (the [::-1] implies that the search starts from the end of the Cs and tspan vectors). We perform the interpolation for both profiles:  

```python
Cs1=N1[:,1]/N1[:,3]   #first integration
Cs2=N2[:,1]/N2[:,3]   #second integration 
tdone1=numpy.interp(0.001,Cs1[::-1],tspan[::-1])  #tdone is time when substrate is depleted
tdone2=numpy.interp(0.001,Cs2[::-1],tspan[::-1])
```

Now we can calculate the productivity in $\frac{cmol}{L\cdot h}$

```python
Cp1=N1[:,2]/N1[:,3]   
Cp2=N2[:,2]/N2[:,3]
P1=max(Cp1)/tdone1   #productivity of first integration
P2=max(Cp2)/tdone2
P2/P1
```

More than twice as fast (2.23x) if there is no substrate inhibition!

What will happen if we feed substrate continuously to the fermenter. We'll start ($t=0$) with a small substrate concentration in the fermenter ($C_S^o = 0.02 \frac{cmol}{L}$) and add a concentrated substrate syrup ($C_S^f=\,10 \frac{cmol}{L}$) at $Q^f=60 \frac{L}{h} $. 

```python
Ks=2   #substrate inhibition is set again
Qf=60  #L/h
Csf=10 #cmol/L 
Cso=0.02                                   
Co=numpy.array([Cxo, Cso, 0, 1])     #[X, S, P, V]  # new initial conditions           
No=Co*Vo

N3,tspan =solve_ferm(40)
```

Wow! What is happening here? Before we think too much, let's pretend we are working in industry where money is the only thing worth thinking about. The profit orientated bioprocess engineer will look at this graph and wonder whether it is more productive than the batch fermenter (N1). To compare he will use the outlet concentration from the batch fermenter ($C_P=2 \frac{cmol}{L}$) and match the concentration on the graph above. The big question is which of the two fermentations get there first. Let's see what the difference in productivity is:

```python
Cp3=N3[:,2]/N3[:,3]   #first integration
t_2=numpy.interp(max(Cp1),Cp3,tspan)  #time to get to 2 cmol/L in fermentation 3
P3=max(Cp1)/t_2
P3/P1
```

Whether he understands or not, he will like the 27% increase in productivity and the fact that the higher product concentrations (sometimes referred to as titre) will be possible with longer fermentation times. But why did we get the productivity increase? Let's go through the motions of chapter 6 part 2 where we plotted the rate profiles. 

```python
Ci1=(N1[:, :3].T/N1[:, 3]).T                  #first 4 elements of N devided by 5th element to get Ci, 
Ci3=(N3[:, :3].T/N3[:, 3]).T
r1=numpy.asarray([r_prime(C) for C in Ci1])
r3=numpy.asarray([r_prime(C) for C in Ci3]) 

rs1=r1[:,1]
rx1=r1[:,0]

rs3=r3[:,1]
rx3=r3[:,0]

#plt.plot(tspan, rp1, color='blue', label='rp(batch)')
#plt.plot(tspan, rp3, color='red',label='rp(fed-batch)')
plt.plot(tspan, rx1, color='green',label='rx(batch)')
plt.plot(tspan, rx3, color='cyan',label='rx(fed-batch)')

plt.legend(loc='best')
plt.ylabel('mu (1/h)') 
plt.xlabel('time (h)') 
plt.show()
```

By plotting $\mu$ against time we gain some insight. Remember that $\mu_{max}=0.4$. It is very clear that the fed-batch growth rates are higher in the beginning. This is directly related to the lower substrate concentration, where the substrate inhibition fraction is less severe. The batch system has a very low initial growth rate since all the substrate was present right from the start (causing high $C_S^o$). Only towards the end did the growth rate climb due to the decrease in $C_S$. The rapid collapse in growth rate at the end is once again the Monod-term kicking in. Let's plot the substrate rate:  

```python
plt.plot(tspan, rs1, color='blue', label='rs(batch)')
plt.plot(tspan, rs3, color='red',label='rs(fed-batch)')

plt.legend(loc='best')
plt.ylabel('rs (cmol S/(cmol X h)') 
plt.xlabel('time (h)') 
plt.show()
```

Very similar to the growth rate plot, but upside down. This indicates that growth rather than maintenance is dominating the production (high $z$). You can plot the $z$ variation with time. Also interesting to note that the rates for the fed-batch fermenter does not drop to zero at the end. What causes this? It is the continual addition of substrate syrup. Unfortunately the addition towards the end is to 'slow' given large amount of biomass in the fermenter ($C_X$). A situation has developed where the response of the microbe (r_prime) is controlled by the rate of substrate addition. Let's have a look at the substrate mole balance:

$$\frac{dN_S}{dt}=Q^F C^F_S-QC_S  + r^{'}_S C_X V $$

We are now only looking at the last part of the fermentation and you will note that similar to $C_S$, $N_S$ remains very close to zero at the end. This implies that $\frac{dN_S}{dt}$ is very small and accordingly:

$$Q^F C^F_S \approx r^{'}_S C_X V $$

The $Q^F C^F_S$ term is dominating (or rate determining) while the $r^{'}_S C_X V$ term just follows (the Monod effect will be severe since $C_S$ is non-zero but very small).

The volumetric depletion rate of substrate clearly show how the '7 fat years' came to an end at 25 hours. The switch indicates  a regime change. In the early regime the microbes were controlling $r_S^{'}$, while the latter regime is characterised by the syrup flow rate $Q^F C^F_S $ controlling $r_S^{'} $.

```python
Cx3=N3[:,0]/N3[:,3]  
rs_vol3=rs3*Cx3

plt.plot(tspan, rs_vol3, color='red',label='rs_vol')
plt.legend(loc='best')
plt.ylabel('Rate cmol/(L h)') 
plt.xlabel('time (h)') 
plt.show()
```

But let's get back to the profit orientated bioprocess engineer. He is not thinking regime changes, he is thinking improvement! How can we get this fermenter to run faster (higher productivity)? Keeping the growth rate ($\mu$) high at all times is the answer, but how do we achieve this? The math of the substrate inhibition suggest that we should keep $C_S$ low in the fermeter (do yourself a favour and plot $\mu$ against $C_S$ to understand how high $C_S$ results in low $\mu$ - you have the magic of Ipython notebooks and it will take only a few minutes). 

So how do we keep the substrate concentration in the fermenter low? We manipulate the feed flowrate ($Q^F$) in order to not add too much or too little. Remember we don't want $C_S$ to be too low, because then the Monod-term will reduce $r^{i}$. A simple idea would be to manipulate $Q^F$ so that more substrate is fed when more biomass is present in the fermenter. We can for example define a linear time dependant function for the inlet flow, such that:

$$Q^{\,f}=k_Q*t$$

We'll have to redefine the D.E. to include the time dependant flowrate:

```python
def dNdt_fun(N,t):
    Cx, Cs, Cp, V = N[0]/N[3],N[1]/N[3],N[2]/N[3],N[3]  #calculating and naming concentration and volume 
    r=r_prime([Cx, Cs, Cp])
    Qf=kQ*t                                             #time dependant flowrate             
    
    return [Qf*Cxf-Q*Cx+(r[0])*Cx*V,
            Qf*Csf-Q*Cs+(r[1])*Cx*V,    
            Qf*Cgf-Q*Cp+(r[2])*Cx*V,                        
            Qf-Q] 
```

We can now define $k_Q$ and integrate the $4^{th}$ fermentation run: 

```python
kQ=6 # l/(h^2)
N4,tspan4 =solve_ferm(30)
```

This looks much better! See how low we manage to keep $C_S$. Surely the productivity will be better, let's compare at $C_P=2 \frac{cmol}{L}$:

```python
Cp4=N4[:,2]/N4[:,3]   #first integration
t_2=numpy.interp(max(Cp1),Cp4,tspan4)  #time to get to 2 cmol/L in fermentation 3
P4=max(Cp1)/t_2
P4/P1
```

Yes, the productivity increase compared to the batch fermentation (run1) is 51%. Is it possible to further increase the productivity? What if we could control $C_S$ at a constant low value? We don't want the value to be too low since the Monod-effect will slow down the production, neither do we want it to be too large. This will imply a fancy flow control scheme where $Q^{\,f}$ is coupled to the production rates in the fermenter. We'll leave this to the profit hungry bioprocess engineer, but we'll help him with the maths.

We need to calculate $Q^{\,f}$ such that  $\frac{dC_S}{dt}=0$. If this is obeyed $C_S$ will remain at the initial value of $C_S^o$. The following manipulation can be performed to obtain the function for $Q^{\,f}$:

$$\frac{dN_S}{dt}=V\frac{dC_S}{dt}+C_S \frac{dV}{dt}  $$

$$\frac{dV}{dt}=Q^{\,f} $$

$$ \frac{dC_s}{dt}=\frac{1}{V} \left[ \frac{dN_S}{dt}-C_S Q^{\,f} \right]=0 $$

$$Q^{\,f}= \frac{1}{C_S} \frac{dN_S}{dt}=\frac{1}{C_S} \left[Q^fC_s^f+r_S^{'}C_XV \right]$$

$$Q^{\,f}= \frac{r_S^{'}C_XV}{C_S-C_S^f} $$

It is relatively easy to put this fancy flowrate function into the D.E.:

```python
def dNdt_fun(N,t):
    Cx, Cs, Cp, V = N[0]/N[3],N[1]/N[3],N[2]/N[3],N[3]  #calculating and naming concentration and volume 
    r=r_prime([Cx, Cs, Cp])
    Qf=r[1]*Cx*V/(Cs-Csf)
    
    return [Qf*Cxf-Q*Cx+(r[0])*Cx*V,
            Qf*Csf-Q*Cs+(r[1])*Cx*V,    
            Qf*Cgf-Q*Cp+(r[2])*Cx*V,                        
            Qf-Q] 
```

```python
Cso=0.02                                   
Co=numpy.array([Cxo, Cso, 0, 1])     #[X, S, P, V]  # new initial conditions           
No=Co*Vo
N5,tspan5 =solve_ferm(20)
```

```python
Cp5=N5[:,2]/N5[:,3]   #first integration
t_2=numpy.interp(max(Cp1),Cp5,tspan5)  #time to get to 2 cmol/L in fermentation 3
P5=max(Cp1)/t_2
P5/P1
```

How is that for an increase? The bioprocess engineer will smile. Let's see what happens when we operate at a higher constant value of $C_S$: 

```python
Cso=0.5                                   
Co=numpy.array([Cxo, Cso, 0, 1])     #[X, S, P, V]  # new initial conditions           
No=Co*Vo
N5,tspan5 =solve_ferm(25)
Cp5=N5[:,2]/N5[:,3]   #first integration
t_2=numpy.interp(max(Cp1),Cp5,tspan5)  #time to get to 2 cmol/L in fermentation 3
P5=max(Cp1)/t_2
P5/P1
```

Not so good as before, why? And if we operate at a much lower $C_S$: 

```python
Cso=0.0005                                   
Co=numpy.array([Cxo, Cso, 0, 1])     #[X, S, P, V]  # new initial conditions           
No=Co*Vo
N5,tspan5 =solve_ferm(40)
Cp5=N5[:,2]/N5[:,3]   #first integration
t_2=numpy.interp(max(Cp1),Cp5,tspan5)  #time to get to 2 cmol/L in fermentation 3
P5=max(Cp1)/t_2
P5/P1
```

Even worse, why? Note that the fermentation times were varied for the different runs, so it is a good idea to always use the returned tspan.
